## Cluster the following eight points (with (x, y) representing locations) into three clusters using K-means

A1(2, 10), A2(2, 5), A3(8, 4), A4(5, 8), A5(7, 5), A6(6, 4), A7(1, 2), A8(4, 9)

```python
import random
import copy
import sys
data={
    '1':[2,10],
    '2':[2,5],
    '3':[8,4],
    '4':[5,8],
    '5':[7,5],
    '6':[6,4],
    '7':[1,2],
    '8':[4,9]
}

centroids={}
distances={}
centroids_clusters={} 

def select_k_centroids(dt,cdroids,k):
    ki = 0
    temp_index_list=[]
    index=None
    while ki<k:
        while True:
            index = random.randint(1,8)
            if index not in temp_index_list:
                temp_index_list.append(index)
                break
        cdroids[str(ki)]= copy.deepcopy(dt[str(index)])
        ki+=1
    del temp_index_list
    return cdroids

def make_place_for_distances(dt,ds,k):
    for d in dt:
        ki=0
        ds[d]=[]
        while ki<k:
            ds[d].append(None)
            ki+=1
    return ds

def find_distance_of_data_points_from_centroids(dt,cdroids,ds):
    for c in cdroids:
        for d in ds:
            ds[d][int(c)] = abs(cdroids[c][0]-dt[d][0])+ abs(cdroids[c][1]-dt[d][1])
    return ds

def find_clusters(dt,cdroids,ds,c_clusters,k):
    min_c=0
    min_v=0
    for c in cdroids:
        c_clusters[c] = []
    for d in dt:
        min_v = ds[d][0]
        min_c = 0
        ki=1
        while ki < k:
            if ds[d][ki]<min_v:
                min_v = ds[d][ki]
                min_c = ki
            ki+=1
        c_clusters[str(min_c)].append(d)
        del ki
    del min_c
    del min_v
    return c_clusters

def take_average_of_clusters_x_and_y(dt,cdroids,c_clusters):
    for c in c_clusters:
        x=0
        y=0
        for cl in c_clusters[c]:
            x+=dt[cl][0]
            y+=dt[cl][1]
        cdroids[c] = [int((x*1.0)/len(c_clusters[c])),int((y*1.0)/len(c_clusters[c]))]
    return cdroids


k = int(input("k = "))
if k>len(data):
    k=len(data)
centroids = select_k_centroids(data,centroids,k)
distances = make_place_for_distances(data,distances,k)

i=0
while True:
    print("**************")
    print("  centroids")
    print("  c# : [x,y]")
    print("**************")
    for c in centroids:
        print("'"+c+"' : ",centroids[c])
    distances = find_distance_of_data_points_from_centroids(data,centroids,distances)
    
    print("\n****************************************")
    print("  p# : [x,y][distances_from_centroids]")
    print("****************************************")
    for d in data:
        print("'"+d+"' : ",data[d],distances[d])
    centroids_clusters = find_clusters(data,centroids,distances,centroids_clusters,k)
    
    print("\n********************************")
    print("        centroid clusters")
    print("  c# : [point1 , point2, ... ]")
    print("********************************")
    for c in centroids_clusters:
        print("'"+c+"' : ",centroids_clusters[c])
    i+=1
    if i==k:
        break
    print("_______________________________________________________________________________________________________________\n")
    centroids = take_average_of_clusters_x_and_y(data,centroids,centroids_clusters)



    

```
