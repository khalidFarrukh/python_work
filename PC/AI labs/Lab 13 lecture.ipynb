# Question
Consider a genetic algorithm (GA) with chromosomes consisting of six genes <b>x:= abcdef</b>, and each gene is a number between 0 and 9. Suppose we have the following population of four chromosomes:
- x1 = 4 3 5 2 1 6
- x2 = 1 7 3 9 6 5 
- X3 = 2 4 8 0 1 2 
- X4 = 9 0 8 1 2 3<br>
and let the fitness function be <b>f(x) = (a + c + e) - (b + d + f)</b>.
- 1. Sort the chromosomes by their fitness.
- 2. Do one-point crossover in the middle between the 1st and 2nd fittest and so on.
- 3. Calculate the fitness of all the new generation.

```python
import copy
import random
import sys
def initialize_population(_p,i_p):
    i=0
    for c in _p:
        _p[c][0]=i_p[i]
        i+=1
    return _p
def fitness_func(c):
    return (int(c[0])+int(c[2])+int(c[4]))-(int(c[1])+int(c[3])+int(c[5]))

def calculate_fitness(_p):
    sum =0
    for c in _p:
        _p[c][1] = abs(fitness_func(_p[c][0]))
        sum+=_p[c][1]
    return _p,sum

def sort_population_based_on_fitness(_p):
    ci=1
    while ci <=len(_p):
        index1 = str(ci)
        cj=ci+1
        while cj <=len(_p):
            index2 = str(cj)
            if _p[index1][1]>_p[index2][1]:
                temp = _p[index1]
                _p[index1] = _p[index2]
                _p[index2] = temp
            cj+=1
        ci+=1
    return _p

def fitness_ratio(_p,sig_sum):
    for c in _p:
        _p[c][2] = float("{:.1f}".format(((_p[c][1]*1.0)/(sig_sum*1.0))*100))
    return _p

            
def Reproduction(_p):
    
    i=1
    i_p=[]
    random.seed()
    length = len(_p)
    while i<length:
        cross1 = _p[str(i)][0]
        i+=1
        cross2 = _p[str(i)][0]
        point= random.randrange(1,4)
        print("single point cross over location = ",point)
        print("parents : ",cross1 , " < - > ",cross2)
        c1a = cross1[:point]
        c1b = cross1[point:]
        c2a = cross2[:point]
        c2b = cross2[point:]
        temp1 =c1a+c2b
        temp2 =c2a+c1b
        print("childs  : ",temp1 , " ***** ",temp2)
        print("\n")
        i_p.append(temp1)
        i_p.append(temp2)
        i+=1
    return i_p




def any_selection_prob_greater_than(inum,_p):
    for c in _p:
        if _p[c][2]>=inum:
            return True
    return False

    

initial_population = ["435216","173965","248012","908123"]

gen = 0
while True:
    _population = {
        '1':[None,None,None,False],
        '2':[None,None,None,False],
        '3':[None,None,None,False],
        '4':[None,None,None,False]
    }
    _population = initialize_population(_population,initial_population)
    _population,sigma_sum = calculate_fitness(_population)
    _population = sort_population_based_on_fitness(_population)
    _population = fitness_ratio(_population,sigma_sum)

    print("\t # :\t chromo\t\tfitness\t\tfitness_ratio")
    for c in _population:
        print("\t",c,":\t",_population[c][0],"\t  ",_population[c][1],"\t\t   ",_population[c][2],end="")
        print()
    print("\n")
    if any_selection_prob_greater_than(50.0,_population)==True:
        break
    initial_population = Reproduction(_population)
    #print(initial_population)
    #initial_population = convert_decimal_to_binary_list_of_strings(_population,length)
    #print(initial_population)

```
